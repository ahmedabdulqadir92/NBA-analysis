توجد أعمدة منفصلة لكل من الفريق المضيف (Home) والفريق الضيف (Away)، مما يسمح بمقارنة الأداء بين الفريقين في نفس المباراة.

## Data Cleaning

في هذه المرحلة تم تنظيف البيانات وتجهيزها للتحليل.

- تم فحص القيم المفقودة (Missing Values)، وتبين وجود عدد قليل منها، لذلك تم حذف الصفوف التي تحتوي على قيم مفقودة.
- تم التأكد من عدم وجود بيانات مكررة (Duplicates).
- تم تحويل نوع عمود التاريخ (GAME_DATE_EST) إلى صيغة تاريخية مناسبة للتحليل.
- تم حذف عمود (GAME_ID) لأنه يمثل معرفًا فريدًا لكل مباراة ولا يحتوي على معلومات مفيدة للتحليل.

بعد عملية التنظيف أصبحت البيانات جاهزة لمرحلة التحليل الاستكشافي (EDA).

## تحليل نقاط الفريق صاحب الأرض (PTS_home)

- متوسط عدد النقاط هو حوالي 103 نقطة، وهذا يدل على أن الأداء الطبيعي للفرق يكون حول هذا الرقم.
- أعلى عدد نقاط مسجل هو 168 نقطة، وهو رقم مرتفع لكنه ممكن في بعض المباريات القوية.
- أقل عدد نقاط هو 36 نقطة، وهو رقم منخفض جدًا وقد يشير إلى حالة نادرة أو مشكلة في البيانات.

بشكل عام، توزيع النقاط يبدو منطقيًا مع وجود بعض القيم المتطرفة.

```python
import plotly.express as px
```

```python
px.histogram(df, x='PTS_home', nbins=50)
```

من خلال تحليل توزيع عدد النقاط للفريق المضيف (PTS_home)، يتضح أن أغلب القيم تتمركز حول 100 نقطة تقريباً، مما يشير إلى أن هذا هو المعدل الطبيعي لتسجيل النقاط في المباريات.

كما نلاحظ أن التوزيع يبدو متوازنًا بدون وجود قيم شاذة أو غير منطقية، وهذا يدل على أن البيانات نظيفة وتعكس الأداء الحقيقي للفرق بشكل جيد.

```python
df[['PTS_home', 'PTS_away']].describe()
```

```python
df['HOME_TEAM_WINS'].value_counts()
```

من خلال تحليل نتائج المباريات، نلاحظ أن الفريق المضيف فاز في 15645 مباراة مقابل 10907 خسارة، مما يشير إلى أن الفريق المضيف يمتلك أفضلية واضحة في الفوز.

هذا يدعم فرضية "أفضلية اللعب على أرض الفريق" (Home Advantage)، حيث يحقق الفريق المضيف نتائج أفضل مقارنة بالفريق الضيف.

```python
fig = px.box(
 df,
 y=['PTS_home', 'PTS_away'],
title='Comparison between Home and Away Points'
 )
fig.update_layout(
    title_x=0.5
 )
fig.show()
```

يوضح الرسم أن توزيع نقاط الفريق المضيف والضيف متقارب، مع تفوق بسيط للفريق المضيف من حيث المتوسط، مما يشير إلى وجود أفضلية محدودة للعب على أرض الفريق.

```python
df[['PTS_home', 'PTS_away', 'HOME_TEAM_WINS']].corr()
```

هناك علاقة إيجابية بين نقاط الفريق صاحب الأرض واحتمالية الفوز، وعلاقة سلبية بين نقاط الفريق الضيف وفوز الفريق صاحب الأرض. كما أن هناك ارتباط متوسط بين نقاط الفريقين، مما يدل على أن المباريات ذات التسجيل العالي تشمل الفريقين معًا

```python
corr = df[['PTS_home', 'PTS_away', 'HOME_TEAM_WINS']].corr()

fig = px.imshow(corr, text_auto=True, title="Correlation Heatmap")
fig.show()
```

يظهر تحليل الارتباط أن تسجيل الفريق صاحب الأرض يؤثر إيجابيًا على احتمالية الفوز، بينما تسجيل الفريق الضيف يؤثر سلبًا. كما توجد علاقة متوسطة بين نقاط الفريقين، مما يدل على أن بعض المباريات تكون ذات طابع هجومي للطرفين.

```python
fig = px.scatter(df, x='PTS_home', y='PTS_away',
 title='Home vs Away Points')
fig.show()
```

"يوضح الرسم وجود علاقة إيجابية بين نقاط الفريق صاحب الأرض والفريق الضيف، مما يشير إلى أن المباريات ذات الطابع الهجومي تؤدي إلى تسجيل نقاط عالية لكلا الفريقين، مع وجود أفضلية بسيطة لصالح الفريق صاحب الأرض."
